```python

import requests
from bs4 import BeautifulSoup
import nltk
import matplotlib.pyplot as plt
```

```python

url = "https://en.wikipedia.org/wiki/Natural_language_processing"
response = requests.get(url)
print("Status code:", response.status_code)
```

```python

soup = BeautifulSoup(response.text, 'html.parser')


paragraphs = soup.find_all('p')


text = ""
for p in paragraphs:
    text += p.get_text()


print(text[:1000])
```

```python

import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords


words = word_tokenize(text.lower())


stop_words = set(stopwords.words('english'))

words_filtered = [word for word in words if word.isalpha() and word not in stop_words]

print("Total words after filtering:", len(words_filtered))
print("First 20 words:", words_filtered[:20])
```

```python
import nltk
nltk.download('punkt_tab')
```

```python
from collections import Counter


word_counts = Counter(words_filtered)


most_common = word_counts.most_common(15)


labels, values = zip(*most_common)


plt.figure(figsize=(12, 6))
plt.bar(labels, values, color='skyblue')
plt.title("Top 15 Most Frequent Words")
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
